```python
import pandas as pd

negative_df = pd.read_csv('../text_model/negative_df.csv', index_col=0)
white_stats_df = pd.read_csv('../data/white_stats.csv', index_col=0)
expression_preds = pd.read_excel('../data/BRCA_predicted_subtypes.xlsx')
text_preds = pd.read_csv('../data/text_predicted_subtypes.csv').rename({'patient_id': 'patients_id'}, axis=1)
base_metadata_df = pd.read_csv('../data/TCGA_breast_patch/metadata.csv', index_col=0)
```

```python
text_preds.head()
```

```python
white_stats_df['patients_id'] = white_stats_df['0'].apply(lambda x: '-'.join(x.split('/')[6].split('-')[0:3])) 
```

```python
len(set(negative_df.index).intersection(set(white_stats_df['patients_id'].unique())))
```

```python
text_preds
```

```python
expression_preds_metadata = pd.merge(white_stats_df[white_stats_df['1'] <= 0.2], text_preds, on='patients_id')
```

```python
expression_preds_metadata
```

```python
expression_preds_metadata = expression_preds_metadata.rename({'0': 'file_name', 'labels': 'label'}, axis=1)[['file_name', 'label']]
expression_preds_metadata['file_name'] = expression_preds_metadata['file_name'].apply(lambda x: x.replace('/mnt/d/uczelnia/magister/', '../data/'))
expression_preds_metadata
```

```python
pd.concat((base_metadata_df, expression_preds_metadata.rename({'0': 'file_name', 'subtype': 'label'}, axis=1)[['file_name', 'label']])).to_csv('../data/TCGA_breast_patch/text_metadata.csv')
```
